```python
import time, pandas as pd, sqlalchemy as sa
from IPython.display import clear_output, display

engine = sa.create_engine("postgresql+psycopg2://mastodon:mastodon@localhost:5433/mastodon")

EVERY    = 3      # seconde(s) entre 2 pulls
LIMIT    = 10     # lignes affich√©es
MIN_NEW  = 1      # nb mini de nouvelles lignes pour redessiner

last_id = None
try:
    i = 0
    while True:
        with engine.connect().execution_options(isolation_level="AUTOCOMMIT") as conn:
            # Dernier id
            mx = pd.read_sql("SELECT max(id) AS mx FROM mastodon_posts;", conn)["mx"].iloc[0]
            if last_id is None:
                last_id = mx

            # Combien de nouvelles lignes ?
            new = (mx or 0) - (last_id or 0)

            if new >= MIN_NEW:
                df = pd.read_sql(sa.text(f"""
                    SELECT id, username, left(content,140) AS preview, ts
                    FROM mastodon_posts
                    ORDER BY id DESC
                    LIMIT :limit
                """), conn, params={"limit": LIMIT})

                clear_output(wait=True)
                print(f"‚è± Refresh #{i} | +{new} nouvelles lignes | id max {mx}")
                display(df)

                last_id = mx
                i += 1

        time.sleep(EVERY)
except KeyboardInterrupt:
    print("‚èπÔ∏è Live view stopped.")
    print("‚èπÔ∏è Live view stopped.")
```
